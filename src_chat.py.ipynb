* Goal: Manage the Q&A memory.

```python
from llama_index.core import Settings

def chat_logic(message, history):
    """
    Standard Chat logic with history.
    'history' is passed automatically by gr.ChatInterface as a list of [user, bot] pairs.
    """
    try:
        full_promptcontext = "You are a legal assistant specialized in German Mietrecht.\n"
        for user_msg, bot_msg in history:
            full_prompt += f"User: {user_msg}\nAssistant: {bot_msg}\n"
        full_prompt += f"User: {message}\nAssistant:"
        resp = Settings.llm.complete(full_prompt)
        return str(resp)
    except Exception as e:
        return f"Ollama Error: {e}"

```
